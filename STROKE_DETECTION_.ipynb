## **Missing Values**

```python
# List of columns to fill missing values with mode
columns_to_fill_with_mode = ['bmi']

# Fill missing values in specified columns with the mode
for column in columns_to_fill_with_mode:
    mode_value = df[column].mode().values[0]  # Calculate the mode value for the column
    df[column].fillna(mode_value, inplace=True)
```

```python
df.isnull().sum()
```

##**Label Encoding**  

```python
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to the categorical columns
df['gender'] = label_encoder.fit_transform(df['gender'])
df['ever_married'] = label_encoder.fit_transform(df['ever_married'])
df['work_type'] = label_encoder.fit_transform(df['work_type'])
df['Residence_type'] = label_encoder.fit_transform(df['Residence_type'])
df['smoking_status'] = label_encoder.fit_transform(df['smoking_status'])

```

```python
df
```

```python
df.drop('id', axis=1, inplace=True)
```

```python
#Splitting the dataset into feature columns and target column
X=df.drop(columns=['stroke'])
Y=df[['stroke']]
```

## Variance & Bias

```python
from sklearn.linear_model import LinearRegression
from mlxtend.evaluate import bias_variance_decomp
#separate into inputs and outputs
data=df.values
X,y=data[:,:-1],data[:,-1]
#split the data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=1)
#define the model
model=LinearRegression()
#estimate bias and variance
mse,bias,var=bias_variance_decomp(model,X_train,y_train,X_test,y_test,loss='mse',num_rounds=200,random_seed=1)
```

```python
#summarize results
print('MSE:%.3f'%mse)
print('Bias:%3f'%bias)
print('Variance:%.3f'%var)
```

# Multiple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing
```

```python
df = pd.read_csv("/content/Real-estate1.csv")
df
```

```python
df.drop('No', inplace = True, axis = 1)
df.head()
```

```python
df.tail()
```

```python
 # dependent var = Y house price of unit area , independent var- rest all 6
```

```python
# creating feature variables
X = df.drop('Y house price of unit area' , axis =1 )
y = df['Y house price of unit area']
```

```python
X
```

```python
y
```

```python
# creating the train test sets

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 101)
```

```python
# creating regression model
model = LinearRegression()

# fitting the model
model.fit(X_train,y_train)

# making predicitions
predictions = model.predict(X_test)
```

```python
# Model Evaluation

print(
    'mean_squared_error : ', mean_squared_error(y_test, predictions))
print(
     'mean_absolute_error : ', mean_absolute_error(y_test, predictions))

# more value - bad model
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, predictions)

print("r2 score is :  ", score)              # value from 0 to 1 , more value - good model
```

```python
model.coef_                # coef=  b1     - depends on the independent feature - so 6                                  y=b0 + b1x

# coeff value with less value can be ommited
# shrinking the coeff to avoid overfit - for this rigg and lasso is used
```

```python
model.intercept_
```

```python
import pandas as pd
import numpy as np

mt = pd.read_csv("/content/mtcars.csv")
mt
```

```python
mt = mt.drop(columns = 'manufacturer', axis = 1)
```

```python
x = mt.drop(columns = 'hp', axis = 1)
y = mt['hp']
```

```python
# splitting the data into train test

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3 , random_state = 4)
```

```python
# fitting the data into the model

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.metrics import r2_score
x_pred_test = lr.predict(x_test)
r2_score(y_test,x_pred_test)
```

```python
# LASSO
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(x_train,y_train)
```

```python
lasso.coef_
```

```python
lasso.intercept_
```

# Ridge

```python
from sklearn.linear_model import Ridge
clf = Ridge(alpha = 2)                      #alpha = 0.001 , 0.005, 1, 2,3 ,4,5
clf.fit(x_train,y_train)
```

```python
x_pred_ridge_test = clf.predict(x_test)
r2_score(y_test, x_pred_ridge_test)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
# alpha = 5

clf = Ridge(alpha = 5)                      #alpha = 0.001 , 0.005, 1, 2,3 ,4,5
clf.fit(x_train,y_train)
x_pred_ridge_test = clf.predict(x_test)
r2_score(y_test, x_pred_ridge_test)
```

```python
# alpha = 4

clf = Ridge(alpha = 4)                      #alpha = 0.001 , 0.005, 1, 2,3 ,4,5
clf.fit(x_train,y_train)
x_pred_ridge_test = clf.predict(x_test)
r2_score(y_test, x_pred_ridge_test)
```

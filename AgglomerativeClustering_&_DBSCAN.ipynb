```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler, normalize
import scipy.cluster.hierarchy as shc
```

```python
X = pd.read_csv("/content/credit.csv")

# Dropping the CUST_ID column from the data
X = X.drop('CUST_ID', axis = 1)

# Handling the missing value
X.fillna(method = 'ffill' , inplace = True)
```

```python
X
```

```python
# Scaling the data so that all the features become comparbale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Normalizing the data so that the data approximately follows a Gaussian Distribution
X_normalized  = normalize(X_scaled)

# Converting the numpy array into a pandas DataFrame
X_normalized  = pd.DataFrame(X_normalized)
```

```python
plt.figure(figsize= (8,8))
plt.title("Visualising the data")
#Dendrogram = shc.dendrogram((shc.linkage(X_principal, method = 'ward')))
Dendrogram = shc.dendrogram((shc.linkage(X_normalized, method = 'ward')))

#  where ever the horizontal line is passing the intersection of that gives the number of clusters
# Eg threshold = 30 , clsuter = 5
```

```python
plt.figure(figsize= (8,8))
plt.title("Visualising the data")
Dendrogram = shc.dendrogram((shc.linkage(X_normalized, method = 'complete')))
```

```python
ac2 = AgglomerativeClustering (n_clusters = 2)
pca = PCA(n_components = 2)
X_principal = pca.fit_transform(X_normalized)
X_principal = pd.DataFrame(X_principal)
X_principal.columns = ['P1','P2']

# Visualizing the Clustering
plt.figure(figsize= (6,6))
plt.scatter (X_principal['P1'],X_principal['P2'],
             c = ac2.fit_predict(X_principal), cmap = 'rainbow')
plt.show()
```

## Aggloromative Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, normalize

from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
```

```python
dataset = pd.read_csv("/content/aggl.csv")

# Dropping the CUST_ID column from the data
X = dataset.iloc[: , [3 , 4]].values
```

```python
# Scaling the data so that all the features become comparbale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Normalizing the data so that the data approximately follows a Gaussian Distribution
X_normalized  = normalize(X_scaled)

# Converting the numpy array into a pandas DataFrame
X_normalized  = pd.DataFrame(X_normalized)
```

```python
Dendrogram = shc.dendrogram((shc.linkage(X_normalized, method = 'ward')))
model = AgglomerativeClustering(n_clusters = 5, affinity= 'euclidean', linkage = 'ward')
model.fit(X)
labels = model.labels_
labels
```

# DBSCAN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, normalize
import scipy.cluster.hierarchy as shc
```

```python
X = pd.read_csv("/content/credit.csv")

# Dropping the CUST_ID column from the data
X = X.drop('CUST_ID', axis = 1)

# Handling the missing value
X.fillna(method = 'ffill' , inplace = True)
```

```python
# Scaling the data so that all the features become comparbale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Normalizing the data so that the data approximately follows a Gaussian Distribution
X_normalized  = normalize(X_scaled)

# Converting the numpy array into a pandas DataFrame
X_normalized  = pd.DataFrame(X_normalized)
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=0.2,min_samples=30)
db.fit(X_normalized)
#db.fit(X_principal)
df=db.labels_
X["labels"]=df
X["labels"].value_counts()
```
